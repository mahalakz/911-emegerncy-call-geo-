```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df.head(3)
```

```python
df['date'] = df['timeStamp'].apply(lambda time: time.date())
df['hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['month'] = df['timeStamp'].apply(lambda time: time.month)
df['year'] = df['timeStamp'].apply(lambda time: time.year)
df['day_of_week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
df.head(3)
```

```python
df['day_of_week_str'] = df['day_of_week'].map(
    {
        0: 'Mon',
        1: 'Tue',
        2: 'Wed',
        3: 'Thu',
        4: 'Fri',
        5: 'Sat',
        6: 'Sun'
    }

)
df['month_str'] = df['month'].map(
    {
        1: 'Jan',
        2: 'Feb',
        3: 'Mar',
        4: 'Apr',
        5: 'May',
        6: 'Jun',
        7: 'Jul',
        8: 'Aug',
        9: 'Sep',
        10: 'Oct',
        11: 'Nov',
        12: 'Dec'
    }
)
```

```python
from datetime import date
df_2016_2019['date_ordinal'] = df_2016_2019['date'].apply(lambda date: date.toordinal())
df_2016_2019.head(3)
```

```python
grouped_by_date = df_2016_2019.groupby('date_ordinal').count()
grouped_by_date.head()
```

```python
grouped_by_date['_date_ordinal'] = grouped_by_date.index
grouped_by_date.head()
```

```python
ax = sns.lmplot(
    data=grouped_by_date,
    x='_date_ordinal',
    y='twp',
    height=4, # appearance
    aspect=9/2, 
    scatter_kws={"s": 10}, # reduce the size of the markers (points)
)
# Appropriate scales for axes
ax.set(xlim=(grouped_by_date['_date_ordinal'].min() - 1, grouped_by_date['_date_ordinal'].max() + 1))
ax.set(ylim=(0, grouped_by_date['twp'].max() + 500))

# Replace ordinal labels with dates on x-axis
ax.set(xlabel='Date', ylabel='Number of calls')
new_labels = [date.fromordinal(int(item)) for item in ax.axes[0,0].get_xticks()]
ax.set(xticklabels=new_labels)

# Rotate labels on x-axis so they don't overlap
ax.axes[0,0].set_xticklabels(ax.axes[0,0].get_xticklabels(), rotation=45, horizontalalignment='right')
```

```python
import folium

def create_montco_map():
    return folium.Map(
        location=[40.2290,-75.3879],
        tiles="Stamen Toner",
        zoom_start=10,
        width = 800,
        height = 600,
    )

montco_map = create_montco_map()
```

```python
# filter dataframe for 1/1/2016
df_1day = df[df['date']<date(year=2016,month=1,day=1)]

#drop NA zip codes
df_1day = df_1day.dropna(subset=['zip'])

#convert zip codes from float to integer now NA values removed
df_1day['zip'] = df_1day['zip'].astype('int16')

# check zip codes
df_1day['zip'].unique()
```

```python
montco_map = create_montco_map()

#for every 911 call on 1/1/2016
for index,row in df[(df['date']==date(year=2016,month=1,day=1)) & (df['hour']<=3)].iterrows():
    
    #set marker depending on category of call
    if row['reason'].strip().lower()=='fire':
        icon = folium.Icon(color='red')
    elif row['reason'].strip().lower()=='traffic':
        icon = folium.Icon(color='orange')
    else:
        icon = folium.Icon(color='green')
        #icon = folium.Icon(icon='glyphicon-plus')
    #define marker and add to map
    folium.Marker(
        location=[row['lat'],row['lng']],
        tooltip=row['desc'],
        icon=icon,
    ).add_to(montco_map),
        
montco_map
```

```python
from folium.plugins import MarkerCluster

montco_map = create_montco_map()

# for every zip code
for zip_code in df_1day['zip'].unique():
    
    df_this_zip = df_1day[df_1day['zip']==zip_code]
    
    # get mean of zip code latitude and longitude for cluster position
    lng_mean = df_this_zip['lng'].mean()
    lat_mean = df_this_zip['lat'].mean()
    # create folium marker cluster
    marker_cluster = MarkerCluster(location=[lng_mean, lat_mean]).add_to(montco_map)
    
    # for every 911 call in this zip code
    for index,row in df_this_zip.iterrows():
        
        #define marker and add to map
        folium.Marker(
            location=[row['lat'],row['lng']],
            #tooltip=row['desc'],
            #icon=icon,
        ).add_to(marker_cluster),
montco_map  
```

```python
from folium.plugins import HeatMap

montco_map = create_montco_map()

locations = []

for index,row in df_1day.iterrows():
    locations.append(
        [
            row['lat'], 
            row['lng'],
            1,
        ]
    )
HeatMap(locations).add_to(montco_map)

montco_map
```
